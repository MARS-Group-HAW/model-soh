```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv

# 1. Connect and Load (Exclude the geometry column to save memory)
load_dotenv("../../.env")

db_url = f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
engine = create_engine(db_url)

df = pd.read_sql(f"SELECT * FROM {os.getenv('DB_NAME')}.{os.getenv('DB_TABLE')} LIMIT 100000", engine)
```

```python
# 2. Visualize energy level, velocity, and remaining distance over time
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
sns.lineplot(data=df, x='tick', y='energylevel')
plt.title('Energy Level Over Time')

plt.subplot(3, 1, 2)
sns.lineplot(data=df, x='tick', y='remainingroutedistancetogoal', color='green')
plt.title('Remaining Distance to Goal Over Time')

plt.subplot(3, 1, 3)
sns.lineplot(data=df, x='tick', y='velocity', color='orange')
plt.title('Velocity Over Time')


plt.tight_layout()
plt.show()
```

```python
from scipy import stats

# Calculate linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(df['tick'], df['energylevel'])
line = slope * df['tick'] + intercept

# Plot results
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='tick', y='energylevel', label='Actual Energy Level')
plt.plot(df['tick'], line, color='red', linestyle='--', label=f'Linear Regression (R²={r_value ** 2:.4f})')

plt.title('Energy Level Over Time with Linear Regression')
plt.xlabel('Tick')
plt.ylabel('Energy Level')
plt.legend()
plt.show()

print(f"R-squared: {r_value ** 2:.4f}")
```
