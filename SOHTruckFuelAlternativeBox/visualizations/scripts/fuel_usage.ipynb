```python
from scipy import stats

# Calculate linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(df_semitruckdriver['tick'], df_semitruckdriver['energylevel'])
line = slope * df_semitruckdriver['tick'] + intercept

# Plot results
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_semitruckdriver, x='tick', y='energylevel', label='Actual Energy Level')
plt.plot(df_semitruckdriver['tick'], line, color='red', linestyle='--', label=f'Linear Regression (R²={r_value ** 2:.4f})')

plt.title('Energy Level Over Time with Linear Regression')
plt.xlabel('Tick')
plt.ylabel('Energy Level')
plt.legend()
plt.show()

print(f"R-squared: {r_value ** 2:.4f}")
```
