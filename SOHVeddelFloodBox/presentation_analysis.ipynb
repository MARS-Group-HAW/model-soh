# Output analysis
Die CSV Dateien werden je Simulation generiert und enthalten unterschiedliche Informationen, die anhand dieses Notebooks ausgewertet werden.

```python
# @title Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Plot all graphs

```python
ax = None 

for i in range(6):
    arrivalGraph = pd.read_csv('.\SOHTravellingBox\out_' +str((i*60))+'_0.02\\arrivalGraph.csv', sep=";")
    deadAgentsGraph = pd.read_csv('.\SOHTravellingBox\out_' +str((i*60))+'_0.02\deadAgentsGraph.csv', sep=";")

    deadAgentsGraph.rename(columns={'Number': ('Number of dead agents - '+str(i)+' minutes')}, inplace=True)
    arrivalGraph.rename(columns={'Number': ('Number of arriving agents - '+str(i)+' minutes')}, inplace=True)  

    if ax is None:
    # @markdown Agents by time (dead and arriving)
        ax = deadAgentsGraph.plot(x='Seconds', y=('Number of dead agents - '+str(i)+' minutes'), legend=False)
    else:
        deadAgentsGraph.plot(ax=ax, x='Seconds', y=('Number of dead agents - '+str(i)+' minutes'), legend=False)

    arrivalGraph.plot(ax=ax, x='Seconds', y=('Number of arriving agents - '+str(i)+' minutes'), grid=True, legend=False)
```

# Plots Minutenabstand

```python
for i in range(6):
    arrivalGraph = pd.read_csv('.\SOHTravellingBox\out_' +str((i*60))+'_0.02\\arrivalGraph.csv', sep=";")
    deadAgentsGraph = pd.read_csv('.\SOHTravellingBox\out_' +str((i*60))+'_0.02\deadAgentsGraph.csv', sep=";")

    deadAgentsGraph.rename(columns={'Number': ('Number of dead agents - '+str(i)+' minutes')}, inplace=True)
    arrivalGraph.rename(columns={'Number': ('Number of arriving agents - '+str(i)+' minutes')}, inplace=True)  

    # @markdown Agents by time (dead and arriving)
    ax = deadAgentsGraph.plot(x='Seconds', y=('Number of dead agents - '+str(i)+' minutes'), title=('Wasseranstieg nach '+ str(i)+' Minuten'))
    arrivalGraph.plot(ax=ax, x='Seconds', y=('Number of arriving agents - '+str(i)+' minutes'), title=('Wasseranstieg nach ' +str(i)+' Minuten'), grid=True)
```
